```python
import pandas as pd
import numpy as np

# Mapping of market names to collateral types
CRYPTO_MARKETS = [
    'eth_wbtc_usdc', 'base_wbtc_usdc', 'base_wbtc_usdt', 'eth_cbbtc_usdc',
    'eth_wbtc_usdt', "eth_wsteth_usdt", 'eth_weth_usdt', 'eth_cbbtc_usdt',
]
PT_MARKETS = [
    "eth_PT-RLP-4SEP2025_usdc", "eth_PT-USD0++-27MAR2025_usdc",
    "eth_PT-USD0++-31OCT2024_usdc", "eth_PT-USDe-25SEP2025_dai",
    "eth_PT-USDe-25SEP2025_usdc", "eth_PT-USDe-25SEP2025_usdt",
    "eth_PT-USDe-27MAR2025_dai", "eth_PT-USDe-27NOV2025_usds",
    "eth_PT-USDe-31JUL2025_dai", "eth_PT-USR-29MAY2025_usdc",
    "eth_PT-csUSDL-31JUL2025_usdc", "eth_PT-lvlUSD-29MAY2025_usdc",
    "eth_PT-mHYPER-20NOV2025_usdc", "eth_PT-reUSD-18DEC2025_usdc",
    "eth_PT-reUSD-25JUN2026_usdc", "eth_PT-sNUSD-5MAR2026_usdc",
    "eth_PT-sdeUSD-1753142406_usdc", "eth_PT-slvlUSD-25SEP2025_usdc",
    "eth_PT-slvlUSD-29MAY2025_usdc", "eth_PT-stcUSD-23JUL2026_usdc",
    "eth_PT-stcUSD-29JAN2026_usdc", "eth_PT-syrupUSDC-28AUG2025_usdc",
    "eth_PT-syrupUSDC-30OCT2025_usdc", "eth_PT-wstUSR-25SEP2025_usdc",
    "eth_PT-wstUSR-27MAR2025_usdc", "eth_PT-wstUSR-27MAR2025_usr",
    "PT-reUSD-25JUN2026_usdc", "PT-siUSD-26MAR2026_usdc",
]
YB_TOKENS = [
    'eth_usr_usdc', 'eth_wsteth_usdc', 'eth_rlp_usdc',
    'eth_usd0++_usdc', 'eth_fxsave_usdc', 'eth_mapollo_usdc',
    'eth_wsrusd_usdc', 'eth_syrupusdc_pyusd', 'eth_susde_pyusd',
    'eth_stcusd_usdc', 'eth_usde_dai', 'eth_mhyper_usdc', 'eth_syrupusdc_usdc',
    'eth_wstusr_usdc','eth_slvlusd_usdc','eth_csusdl_usdc', 'eth_mF-ONE_usdc', 'eth_reusd_usdc',
    'eth_siusd_usdc', 'eth_sdeusd_usdc'
]

def classify_market(name):
    if name in CRYPTO_MARKETS:
        return "Crypto"
    elif name in PT_MARKETS:
        return "PT"
    elif name in YB_TOKENS:
        return "YB"
    else:
        return "Other"

# ------------------------------------------------------------
# Read data (adjust file path as needed)
df = pd.read_csv(SPIKES_PATH)  # <--- replace with actual filename

# Basic cleaning / type conversion
df['spike_trigger_datetime'] = pd.to_datetime(df['spike_trigger_datetime'])
df['spike_recovery_datetime'] = pd.to_datetime(df['spike_recovery_datetime'])
df['duration_hours'] = df['spike_duration_seconds'] / 3600.0

# Classify markets
df['market_type'] = df['market_name'].apply(classify_market)

# ------------------------------------------------------------
# Remove outliers: spikes with duration > 99th percentile
dur_99 = df['duration_hours'].quantile(0.995)
print(f"99-й перцентиль длительности: {dur_99:.2f} ч")
n_before = len(df)
df = df[df['duration_hours'] <= dur_99]
n_after = len(df)
print(f"Удалено выбросов: {n_before - n_after} (оставлено {n_after} эпизодов)\n")

# ------------------------------------------------------------
# 1. Overall sample characteristics
print("="*60)
print("ОБЩАЯ ХАРАКТЕРИСТИКА ВЫБОРКИ ЭПИЗОДОВ (без выбросов по длительности)")
print("="*60)
total_spikes = len(df)
n_markets = df['market_name'].nunique()
print(f"Всего эпизодов (спайков): {total_spikes}")
print(f"Количество уникальных рынков: {n_markets}")

# Episodes by market type
type_counts = df['market_type'].value_counts()
type_pct = (type_counts / total_spikes * 100).round(1)
print("\nРаспределение по типам рынков (коллатерала):")
for t in ['Crypto', 'PT', 'YB']:
    cnt = type_counts.get(t, 0)
    pct = type_pct.get(t, 0.0)
    print(f"  {t}: {cnt} эпизодов ({pct}%)")

# ------------------------------------------------------------
# 2. Duration statistics
print("\n" + "="*60)
print("СТАТИСТИКИ ДЛИТЕЛЬНОСТИ ЭПИЗОДОВ (в часах)")
print("="*60)

dur = df['duration_hours']
print("Общая выборка:")
print(f"  Средняя:  {dur.mean():.2f} ч")
print(f"  Медиана:  {dur.median():.2f} ч")
print(f"  Мин.:     {dur.min():.2f} ч")
print(f"  Макс.:    {dur.max():.2f} ч")
print(f"  Стандартное отклонение: {dur.std():.2f} ч")

# By market type
print("\nПо типам рынков:")
dur_stats = df.groupby('market_type')['duration_hours'].agg(
    Mean='mean', Median='median', Min='min', Max='max', Std='std', Count='count'
).round(2)
print(dur_stats.to_string())

# ------------------------------------------------------------
# 3. Distribution of durations (buckets)
bins = [0, 1, 6, 24, np.inf]
labels = ['≤ 1 ч', '1–6 ч', '6–24 ч', '> 24 ч']
df['dur_bucket'] = pd.cut(df['duration_hours'], bins=bins, labels=labels, right=True)
bucket_counts = df['dur_bucket'].value_counts().sort_index()
bucket_pct = (bucket_counts / total_spikes * 100).round(1)

print("\n" + "="*60)
print("РАСПРЕДЕЛЕНИЕ ДЛИТЕЛЬНОСТЕЙ (ГРУППЫ)")
print("="*60)
print("Общая выборка:")
for bucket in labels:
    cnt = bucket_counts.get(bucket, 0)
    pct = bucket_pct.get(bucket, 0.0)
    print(f"  {bucket}: {cnt} эпизодов ({pct}%)")

# By market type (optional, for Table 3)
bucket_by_type = df.groupby('market_type')['dur_bucket'].value_counts().unstack(fill_value=0)
bucket_by_type_pct = bucket_by_type.div(bucket_by_type.sum(axis=1), axis=0) * 100
print("\nПо типам рынков (количество):")
print(bucket_by_type.to_string())
print("\nПо типам рынков (проценты):")
print(bucket_by_type_pct.round(1).to_string())

# ------------------------------------------------------------
# 4. Trigger event types – aggregate by first event
print("\n" + "="*60)
print("ТРИГГЕРЫ ЭПИЗОДОВ (ТИПЫ СОБЫТИЙ, агрегировано по первому событию)")
print("="*60)

# Extract first event from the comma-separated string
df['first_trigger'] = df['trigger_event_types'].str.split(',').str[0].str.strip()
trigger_counts = df['first_trigger'].value_counts()
trigger_pct = (trigger_counts / total_spikes * 100).round(1)
trigger_table = pd.DataFrame({
    'Количество': trigger_counts,
    'Доля, %': trigger_pct
})
print(trigger_table.to_string())

# ------------------------------------------------------------
# 5. Supply and Repay events within spikes
print("\n" + "="*60)
print("СОБЫТИЯ MarketSupply И MarketRepay ВНУТРИ ЭПИЗОДОВ")
print("="*60)

# Check if we have explicit column for supply events
if 'n_supply_events_in_spike' in df.columns:
    avg_supply = df['n_supply_events_in_spike'].mean()
    print(f"Среднее число MarketSupply за эпизод: {avg_supply:.2f}")
else:
    # Approximate supply events as total minus repay (with caution)
    df['n_supply_proxy'] = df['n_total_events_in_spike'] - df['n_repay_events_in_spike']
    avg_supply = df['n_supply_proxy'].mean()
    print("Колонка n_supply_events_in_spike отсутствует.")
    print("Аппроксимация: среднее число прочих событий (включая MarketSupply) = n_total - n_repay")
    print(f"Среднее число MarketSupply (приблизительно): {avg_supply:.2f}")

if 'n_repay_events_in_spike' in df.columns:
    avg_repay = df['n_repay_events_in_spike'].mean()
    print(f"Среднее число MarketRepay за эпизод: {avg_repay:.2f}")
else:
    avg_repay = float('nan')
    print("Колонка n_repay_events_in_spike отсутствует.")

# ------------------------------------------------------------
# 6. Preliminary observations (correlations with duration)
print("\n" + "="*60)
print("КОРРЕЛЯЦИИ ДЛИТЕЛЬНОСТИ (на очищенных данных)")
print("="*60)
corr_vars = ['duration_hours', 'peak_utilization', 'market_n_users',
             'spike_magnitude_util_delta', 'n_total_events_in_spike']
corr_matrix = df[corr_vars].corr().round(3)
print(corr_matrix)

# Mean duration by first trigger type
mean_dur_by_trigger = df.groupby('first_trigger')['duration_hours'].mean().round(2)
print("\nСредняя длительность по типу первого триггера (часы):")
print(mean_dur_by_trigger.to_string())

from scipy.stats import mannwhitneyu

df['has_close_event'] = (
    (df['n_repay_events_in_spike'] > 0) |
    (df['trigger_event_types'].str.contains('MarketWithdrawCollateral', na=False))
)

# Средняя длительность по market_type и наличию close-события
dur_close_stats = df.groupby(['market_type', 'has_close_event'])['duration_hours'].agg(
    Mean='mean', Count='count'
).round(2)
print("\nСредняя длительность спайков с событиями закрытия долга (и без):")
print(dur_close_stats)

# Тест Манна–Уитни для каждого market_type
print("\nP-value Манна–Уитни (H1: длительность с закрытием > без закрытия):")
for mtype in ['Crypto', 'PT', 'YB']:
    sub = df[df['market_type'] == mtype]
    dur_with = sub[sub['has_close_event']]['duration_hours']
    dur_without = sub[~sub['has_close_event']]['duration_hours']
    if len(dur_with) == 0 or len(dur_without) == 0:
        p = float('nan')
    else:
        stat, p = mannwhitneyu(dur_with, dur_without, alternative='greater')
    print(f"  {mtype}: p-value = {p:.4e}   (N_with={len(dur_with)}, N_without={len(dur_without)})")

```
