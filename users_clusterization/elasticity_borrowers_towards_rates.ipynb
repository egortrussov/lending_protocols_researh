```python
def hourly_to_daily(hourly_df, apy_df, n_hours=24):
    """
    Aggregate hourly market data into bins of n_hours (default 24).
    Then merge with the closest APY value (by timestamp) and compute rate spread.

    Parameters:
    hourly_df : DataFrame with columns: timestamp, datetime, total_supply, total_borrow,
                utilization, borrow_rate, supply_rate, collateral_price, loan_asset_price, asset_price
    apy_df    : DataFrame with columns: timestamp, apy
    n_hours   : number of hours per aggregation bin (default 24)

    Returns:
    daily_df  : DataFrame with same columns as original daily output plus 'apy' and 'rate_spread'.
    """
    df = hourly_df.copy()
    # Create bin start timestamp (floor division)
    bin_seconds = n_hours * 3600
    df['bin_ts'] = (df['timestamp'] // bin_seconds) * bin_seconds
    df['bin_dt'] = pd.to_datetime(df['bin_ts'], unit='s')

    # Aggregate within each bin
    daily = df.groupby('bin_ts').agg(
        total_supply=('total_supply', 'last'),
        total_borrow=('total_borrow', 'last'),
        utilization=('utilization', 'mean'),
        borrow_rate=('borrow_rate', 'mean'),
        supply_rate=('supply_rate', 'mean'),
        collateral_price=('collateral_price', 'last'),
        loan_asset_price=('loan_asset_price', 'last'),
        asset_price=('asset_price', 'last'),
        datetime=('datetime', 'last')   # representative datetime of the bin
    ).reset_index().rename(columns={'bin_ts': 'timestamp'})

    # Sort for merge_asof
    daily = daily.sort_values('timestamp')
    apy_sorted = apy_df[['timestamp', 'apy']].dropna().sort_values('timestamp')

    # Merge nearest APY
    daily = pd.merge_asof(daily, apy_sorted, on='timestamp', direction='nearest')

    # Compute spread
    daily['rate_spread'] = daily['apy'] - daily['borrow_rate']

    # Optionally drop helper columns if any
    return daily


daily_market_features = hourly_to_daily(market_df, apy_df=apy_df)
daily_market_features.head()
```

```python
import pandas as pd
import numpy as np

def prepare_event(df, event_time, pre=48, post=72):
    t = pd.to_datetime(event_time)
    x = df.copy()
    x["datetime"] = pd.to_datetime(x["datetime"])
    x = x[(x.datetime >= t - pd.Timedelta(hours=pre)) & (x.datetime <= t + pd.Timedelta(hours=post))].copy()
    x["h"] = ((x.datetime - t).dt.total_seconds() / 3600).astype(int)
    b0 = x.loc[x.h == 0, "total_borrow"].iloc[0]
    r0 = x.loc[x.h == 0, "borrow_rate"].iloc[0]
    x["borrow_change"] = (x.total_borrow - b0) / b0
    x["rate_change"] = x.borrow_rate - r0
    return x
def xcorr(x, y, max_lag=48, max_lag_left=24):
    lags = np.arange(-max_lag_left, max_lag + 1)
    vals = []
    for l in lags:
        if l < 0:
            c = np.corrcoef(x[:l], y[-l:])[0,1]
        elif l > 0:
            c = np.corrcoef(x[l:], y[:-l])[0,1]
        else:
            c = np.corrcoef(x, y)[0,1]
        vals.append(c)
    return pd.DataFrame({"lag": lags, "corr": vals})

import matplotlib.pyplot as plt

def plot_event(x):
    fig, ax1 = plt.subplots(figsize=(10,5))
    ax1.plot(x.h, x.borrow_rate, lw=2)
    ax1.axvline(0, ls=":")
    ax1.set_xlabel("Hours from spike")
    ax1.set_ylabel("Borrow rate")
    ax2 = ax1.twinx()
    ax2.plot(x.h, x.borrow_change, lw=2, ls="--")
    ax2.set_ylabel("Borrow change")
    plt.title("Borrower response to rate spike")
    plt.tight_layout()

def plot_xcorr(c):
    plt.figure(figsize=(8,4))
    plt.plot(c["lag"], c["corr"], lw=2)
    plt.axvline(0, ls=":")
    plt.xlabel("Lag (hours)")
    plt.ylabel("Correlation")
    plt.title("Cross-correlation")
    plt.tight_layout()


df_prep = hourly_to_daily(market_df, apy_df=apy_df, n_hours=24)

event_df = prepare_event(
    df_prep,
    event_time="2026-04-18"
)

corr_df = xcorr(
    event_df.rate_change.values,
    event_df.borrow_change.values,
    max_lag=24*3
)

plot_xcorr(corr_df)
```
